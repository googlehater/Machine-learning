```python
def  replace_outliers_nan(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    column.loc[(column < lower_bound) | (column > upper_bound)] = np.nan
    return column
```

```python
df_new['powerPS'] = replace_outliers_nan(df_new['powerPS'])
```

```python
imputer = KNNImputer(n_neighbors=10)
df_new[['powerPS']] = imputer.fit_transform(df_new[['powerPS']])
```

**Задание 8.**	Определите коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и мощностью двигателя. Оцените значение p-value. Постройте графики плотностей распределения признаков (kdeplot или displot).

Коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и мощностью двигателя

Вычисление коэффициентов 

```python
# принимают x и y
pearson_corr, pearson_pval = pearsonr(df_new['price'], df_new['powerPS'])
spearman_coorr, spearman_pval = spearmanr(df_new['price'], df_new['powerPS'])
```

```python
print(f'Корреляция Пирсона:  {pearson_corr:.2f}, p-value: {pearson_pval:.10f}\nКорреляция Спирмена: {spearman_coorr:.2f}, p-value: {spearman_pval:.10f}')
```

Как видно, p-value показывает ноль, однако это не значит, что оно равно нулю, p-value очень близко к нулю. Это указывает на статистичыескую значимость связи. 

Коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и годом регистрации

```python
pearson_corr1, pearson_pval1 = pearsonr(df_new['price'], df_new['yearOfRegistration'])
spearman_coorr1, spearman_pval1 = spearmanr(df_new['price'], df_new['yearOfRegistration'])
```

```python
print(f'Корреляция Пирсона:  {pearson_corr1:.2f}, p-value: {pearson_pval1:.100f}\nКорреляция Спирмена: {spearman_coorr1:.2f}, p-value: {spearman_pval1:.100f}')
```
